```python
Mtot = 70.0
q = 4.7

m1 = Mtot*q/(1+q)*lal.MSUN_SI
m2 = Mtot/(1+q)*lal.MSUN_SI

s1x = 0.3
s1y = 0.
s1z = 0.

s2x = 0.
s2y = -0.5
s2z = -0.6

distance = 100*lal.PC_SI*1e6
incl = np.pi/3

fmin =20.0
fref = 20.0
phiref = -0.8

lalDict = lal.CreateDict()
lalDict2 = lal.CreateDict()

modelist = lalsim.SimInspiralCreateModeArray()
lalsim.SimInspiralModeArrayActivateMode(modelist, 2, 2)
lalsim.SimInspiralModeArrayActivateMode(modelist, 2, -2)
lalsim.SimInspiralModeArrayActivateMode(modelist, 2, 1)
lalsim.SimInspiralModeArrayActivateMode(modelist, 2, -1)
lalsim.SimInspiralModeArrayActivateMode(modelist, 3, 3)
lalsim.SimInspiralModeArrayActivateMode(modelist, 3, -3)
lalsim.SimInspiralModeArrayActivateMode(modelist, 4, 4)
lalsim.SimInspiralModeArrayActivateMode(modelist, 4, -4)
#lalsim.SimInspiralModeArrayActivateMode(modelist, 5, 5)
#lalsim.SimInspiralModeArrayActivateMode(modelist, 5, -5)
lalsim.SimInspiralWaveformParamsInsertModeArray(lalDict, modelist)
```

```python
hpSEOB,hcSEOB = lalsim.SimInspiralChooseTDWaveform(m1,m2,s1x,s1y,s1z,s2x,s2y,s2z,distance,incl,phiref,0,0,0,1./4096,fmin,fref,lalDict2,lalsim.SEOBNRv4PHM)
epochSEOB = hpSEOB.epoch.gpsSeconds + hpSEOB.epoch.gpsNanoSeconds/1e9
timesSEOB = np.arange( hpSEOB.data.length ) * hpSEOB.deltaT + epochSEOB
hSEOB = hpSEOB.data.data -1j*hcSEOB.data.data

# hpN,hcN = lalsim.SimInspiralChooseTDWaveform(m1,m2,s1x,s1y,s1z,s2x,s2y,s2z,distance,incl,phiref,0,0,0,1./4096,fmin,fref,lalDict2,lalsim.NRSur7dq4)
# epochN = hpN.epoch.gpsSeconds + hpN.epoch.gpsNanoSeconds/1e9
# timesN = np.arange( hpN.data.length ) * hpN.deltaT + epochN
# hN = hpN.data.data -1j*hcN.data.data
```
